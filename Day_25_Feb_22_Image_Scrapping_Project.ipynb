```python
import requests
from bs4 import BeautifulSoup
from urllib.request import urlopen
import logging
import os
```

```python
save_dir = "images/"
if not os.path.exists(save_dir):
    os.makedirs(save_dir)
```

```python
# adding header so that google doesn't block us as we are scraping images continuously from web
headers = {"User-Agent": "Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/58.0.3029.110 Safari/537.36"}
```

```python
query = "elon musk"
response = requests.get(f"https://www.google.com/search?q={query}&sxsrf=AJOqlzVhYcQCeO5yKJC-JF3Y1YR1NRJfbA:1677179371116&source=lnms&tbm=isch&sa=X&ved=2ahUKEwi2go-FrKz9AhX3DrcAHe1yA4IQ_AUoAnoECAEQBA&biw=1319&bih=617&dpr=1")
```

```python
response
```

```python
soup = BeautifulSoup(response.content, "html.parser")
```

```python
# scrapping links for all the images
image_tags = soup.find_all("img")
```

```python
# deleting the first record since it is not an image
del image_tags[0]
```

```python
# total no. of images
len(image_tags)
```

```python
image_tags
```

```python
# for bulk upload og data in mongoDB creating a dictionary

image_data_mongo = []
for i in image_tags:
    image_url = i['src']  # getting the url links from tag
    image_data = requests.get(image_url).content   # getting all the contents of the url link
    mydict = {"index" : image_url, "image" : image_data}
    image_data_mongo.append(mydict)
    
    # storing the images in our local system/directory i.e "images/" as well that we created newly as elon musk_01.jpg, elon musk_02.jpg, etc
    with open(os.path.join(save_dir, f"{query}_{image_tags.index(i)}.jpg"), "wb") as f:
        f.write(image_data)
```
